```python
#Importing required libraries
import pandas as pd
from sqlalchemy import create_engine
```

### Importing the cleaned csv data into pandas dataframe and checking data appropriateness

```python
crime_file = "crime2013_2019.csv"
crime_df = pd.read_csv(crime_file,index_col = 0)
crime_df.head()
```

#### Splitting Date strings into month, day, and year for easier mysql import

```python
crime_df["Month"]=crime_df["Date"].str.split('/', expand=False).str[0].astype(str).astype(int)
crime_df["Day"]=crime_df["Date"].str.split('/', expand=False).str[1].astype(str).astype(int)
crime_df["Year"]=crime_df["Date"].str.split('/', expand=False).str[2].astype(str).astype(int)
```

```python
crime_df.head()
```

```python
crime_df.columns
```

```python
#Importing flushot data as dataframe
flushot_file = "cleaned_flu_df.csv"
flushot_df = pd.read_csv(flushot_file, index_col = 0)
flushot_df.head()
```

```python
flushot_df["Year"]=flushot_df["Season"].str.split('-', expand=False).str[0].astype(str).astype(int)
flushot_df.head()
```

```python
#Importing census data as dataframe
census_file = "cleaned_census_df.csv"
census_df = pd.read_csv(census_file, index_col = 0)
census_df.head()
```

```python
census_df.columns
```

```python
census_df["total_population"] = census_df["WHITE"] + census_df["HISP"] + census_df["BLACK"] + census_df["ASIAN"] + census_df["OTHER"]
census_df["percent_unemployed"] = round((1- (census_df["EMP"] /census_df["IN_LBFRC"] ))*100 ,2)
census_df["percent_working"] = round((census_df["EMP"] /census_df["total_population"] )*100, 2)
```

```python
census_df.head()
```

## Connect to the database

```python
# Creating a connection string and engine for mysql local database 
# replace {password} with your password for mysql database connection
connection_string = 'mysql://root:[password]@localhost:3306/shot_chicago'
engine = create_engine(connection_string)
```

```python
# Checking connection using dummy tablename
engine.table_names()
```

# Creating tables and inserting data for the dataframe in sql 

```python
crime_df.to_sql(name='chicrime', con=engine, if_exists='replace', index=False)
```

```python
flushot_df.to_sql(name='flushot', con=engine, if_exists='replace', index=False)
```

```python
census_df.to_sql(name='census', con=engine, if_exists='replace', index=False)
```

Query tables for combined data using community area number

```python
query = """
SELECT
*
FROM census A
INNER JOIN flushot B on A.community_area_number = B.community_area_number
"""

pd.read_sql_query(query, con=engine).head()
```
